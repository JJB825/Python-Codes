```python
df = pd.DataFrame(np.random.randint(low = 3, high = 20, size = (3,2)), index = np.arange(1,4), columns = list(np.arange(1,3)))
df.head()
```

```python
df.describe()
```

```python
print("Mean :-")
print(df.mean(axis = 0))
print()
print("Count :-")
print(df.count(axis = 1))
print()
print("Min :-")
print(df.min(axis = 0))
print()
print("Max :-")
print(df.max(axis = 1))
print()
print("Median :-")
print(df.median(axis = 1))
print()
print("Standard Deviation :-")
print(df.median(axis = 0))
print()
```

```python
df.corr(method = "pearson") # pearson gives standard correlation coeffient
```

```python
df1 = pd.DataFrame(np.random.randint(low = 2, high = 23, size = (3,2)), index = ["First", "Second", "Third"], columns = list("AB"))
df2 = pd.DataFrame(np.random.randint(low = 3, high = 24, size = (3,2)), index = ["First", "Second", "Third"], columns = list("AB"))
with pd.ExcelWriter("Output.xlsx") as writer:
  df1.to_excel(writer, sheet_name = "Sheet1")
  df2.to_excel(writer, sheet_name = "Sheet2")
```

```python
df3 = pd.DataFrame(np.random.randint(low = 2, high = 23, size = (3,2)), index = ["First", "Second", "Third"], columns = list("AB"))
df4 = pd.DataFrame(np.random.randint(low = 3, high = 24, size = (3,2)), index = ["First", "Second", "Third"], columns = list("AB"))
with pd.ExcelWriter("Output.xlsx", mode = 'a') as writer:
  df3.to_excel(writer, sheet_name = "Sheet3", index = False)
  df4.to_excel(writer, sheet_name = "Sheet4", index = False)
```
