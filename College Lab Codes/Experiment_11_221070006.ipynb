```python
# To convert jpg/jpeg images in png format
from PIL import Image as img
import os
for f in os.listdir('.'):
  if f.endswith('.jpg'):
    i = img.open(f)
    name, ext = os.path.splitext(f)
    i.save('{}.png'.format(name))
```

```python
# To convert pictures into size 100 px
size = (100,100)

for f in os.listdir('.'):
  if f.endswith('.jpg'):
    i = img.open(f)
    name, ext = os.path.splitext(f)
    i.thumbnail(size)
    i.save('{}_100{}'.format(name, ext))
```

```python
# To rotate any image
# Positive angle for anticlockwise rotation and Negative angle for clockwise rotation
from PIL import Image as img
i = img.open('I_100.jpg')
rotated_left = i.rotate(90)
rotated_right = i.rotate(-90)
display(rotated_left)
display(rotated_right)
```

```python
# To make an image black and white
i = img.open("H_100.jpg")
black_white = i.convert(mode = "L")
display(black_white)
```

```python
#To make an image blur
from PIL import ImageFilter as img_f
i = img.open('S_100.jpg')
blurred = i.filter(img_f.GaussianBlur())
display(blurred)
```

```python
img1 = img.open('jaguar.jpg')
# In crop the parameter taken is the dimensions of box to be cropped.
# The first two numbers are x, y coordinates of left-most point.
# The next two numbers are width and height of required image
img1 = img1.crop((44,93,435,245))
img1 = img1.resize((400,200))
display(img1)
```

```python
img2 = img.open('J_100.jpg')
img.Image.paste(img1, img2, (160, 60))
display(img1)
```

```python
i = img.open("N_100.jpg")
flippedVertical = i.transpose(method = img.FLIP_TOP_BOTTOM)
flippedHorizontal = i.transpose(method = img.FLIP_LEFT_RIGHT)
display(flippedVertical)
print()
display(flippedHorizontal)
```

```python
image_list = ["J_100.jpg", "I_100.jpg", "S_100.jpg", "H_100.jpg", "N_100.jpg", "U_100.jpg"]
name = img.new('RGB', (600,100))
a = 0
for image in image_list:
  i = img.open(image)
  i = i.resize((100,100))
  name.paste(i, (a*100, 0))
  a += 1

name.save("MyName.jpg")
display(name)
```
