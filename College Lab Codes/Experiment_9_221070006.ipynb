```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/earthquake_data.csv')
```

```python
data
```

```python
data = data.head(5)
```

```python
df = pd.DataFrame(data, columns=["magnitude", "cdi", "mmi"])
```

```python
plt.figure(figsize=(5,5))
plt.plot(df["magnitude"], df["cdi"], c = 'Lime', marker = 'o', ms = 10, mfc = 'Crimson', label = "CDI")
plt.plot(df["magnitude"], df["mmi"], c = 'Cyan', marker = 'o', ms = 10, mfc = 'Indigo', label = "MMI")

plt.xlabel('Magnitude')
plt.ylabel('Intensity')
plt.title('Magnitude vs Intensity')
plt.legend()

plt.show()
```

```python
data = pd.read_csv('/content/earthquake_data.csv')
data = data.tail(10)
df = pd.DataFrame(data,columns=["depth","latitude","longitude"])
plt.figure(figsize=(5,5))
plt.scatter(df["depth"], df["latitude"], color = 'Lime', marker = 'o', label = "Latitude")
plt.scatter(df["depth"], df["longitude"], color = 'Cyan', marker = 'o', label = "Longitude")

plt.xlabel('Depth')
plt.ylabel('Location')
plt.title('Depth vs Location')
plt.legend()
plt.grid()
plt.show()
```

```python
data = pd.read_csv('/content/earthquake_1995-2023.csv')
```

```python
data
```

```python
data = data.head(5)
df = pd.DataFrame(data, columns=["magnitude", "cdi", "mmi"])
bar_width = 0.35
x = np.arange(len(df["magnitude"]))
plt.bar(x-bar_width/2, df["cdi"], color = "DarkMagenta", width = bar_width, label = "CDI")
plt.bar(x+bar_width/2, df["mmi"], color = "DarkOrange", width = bar_width, label = "MMI")
plt.xlabel("Magnitude")
plt.ylabel("Intensity")
plt.legend()
plt.show()
```

```python
data = pd.read_csv('/content/sample_data/california_housing_test.csv')
```

```python
data
```

```python
from random import randint
row = randint(0,2999)
df = pd.DataFrame(data)
rows = df.loc[row,["total_rooms","total_bedrooms","households"]]
row_values = rows.values
labelvals = ["total rooms","total bedrooms","households"]
plt.figure(figsize = (5,5))
plt.title('Home Records')
plt.pie(row_values, labels = labelvals, startangle = 90)
plt.show()
```

```python
plt.hist(df["population"], bins=20, range=(0,9000), color = "Gold")
plt.xlabel('Value')
plt.ylabel('Population')
plt.title("Population Records")
plt.figure(figsize = (5,5))
plt.show()
```
