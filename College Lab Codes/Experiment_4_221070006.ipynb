```python
app = {}  # Initialize an empty dictionary
name = ""  # Initialize the name variable outside the loop

while True:
    print('Choose according to option number : \n1. Add Application\n2. View Application\n3. Delete Application\n4. Exit\n')
    choice = int(input('Enter Choice : '))

    if choice == 1:
        name = input('Enter your name: ')
        num = input('Enter your mobile number: ')
        try:
            num = int(num)
        except ValueError:
            print('Enter valid input')
            continue

        if name in app:
            app[name].update({"Mobile Number": num})
        else:
            app[name] = {"Mobile Number": num}

        email = input('Enter your email address: ')
        app[name].update({"Email Address": email})

        taluka = input('Enter your name of Taluka: ')
        app[name]["Address"] = {"Taluka": taluka}

        city = input('Enter your name of city: ')
        app[name]["Address"].update({"City": city})

        pincode = input('Enter your Pincode: ')
        try:
            pincode = int(pincode)
        except ValueError:
            print('Enter valid input')
            continue

        app[name]["Address"].update({"Pincode": pincode})
        print('Application added successfully.')

    elif choice == 2:
        if name in app:
            print('Name:', name)
            print('Mobile Number:', app[name]["Mobile Number"])
            print('Email Address:', app[name]["Email Address"])
            print('Address:', app[name]["Address"]["Taluka"], app[name]["Address"]["City"], '-', app[name]["Address"]["Pincode"])
        else:
            print('Application not found.')

    elif choice == 3:
        if name in app:
            delete = input('Enter what has to be deleted (Mobile Number, Email Address, Address): ')
            if delete in app[name]:
                del app[name][delete]
                print(f'{delete} deleted successfully.')
                print(app)
            else:
                print(f'{delete} not found.')

        else:
            print('Application not found.')

    elif choice == 4:
        print('Thank You')
        break

    else:
        print('Invalid choice. Please choose a valid option.')

    ch = input('Do you want to continue? (y/n): ')
    if ch != 'y':
        print('Thank You')
        break
```
